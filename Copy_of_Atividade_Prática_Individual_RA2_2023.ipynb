```python
Q1_1exp = ((x**2)-4)*((x**3)-7*x+1)
sy.plot(Q1_1exp, label ='f(x) = (x^2 - 4)(x^3 - 7x + 1)', legend=True)
```

# (B1): 
## $f(x) = (x^2 - 4)(x^3 - 7x + 1)$
## $ g(x)=(x^2 -2)$
## $ h(x)=(x^3 - 7x + 1) $
## $ g'(x) = (2x) $
## $ h'(x) = (3x^2 -7) $
## $ f'(x)=g(x)h'(x) + h(x)g'(x)$
## $ f'(x)=(x^2 - 4)(3x^2 -7) + (x^3 - 7x + 1)(2x) $
## O angulo de inclinação da tangente quando x = 5 é
## $ 1/\tan(f'(5)) $

```python
b1_a = (x**2)-4 # parte A da função
b1_b = (x**3)-(7*x)+1 # parte B da função
b1_aD = 2*x # derivado da parte A
b1_bD = (3*x**2)-7 # derivado da parte B
```

```python
b1Derivative = (b1_a*b1_bD)+(b1_b*b1_aD)
display((b1_a*b1_bD)+(b1_b*b1_aD))
```

```python
sy.tan(b1Derivative.subs(x,5))**-1
```

```python
Q1_2exp = sy.sin(x**2)*(sy.cos(2*x))
sy.plot(Q1_2exp, label ='f(x) = sin(x^2)cos(2x)', legend=True)
```

## (B2):
## $ f(x) = \sin(x^2)\cos(2x)$
## $ g(x)= \sin(x^2) $
## $ h(x)= \cos(2x)$
## $ g'(x)= 2x\cos(x^2)$
## $ h'(x)= -2 \sin(2x)$
## $ f'(x) = (\sin(x^2))(-2 \sin(2x)) + (\cos(2x))(2x\cos(x^2))$
## O angulo de inclinação da tangente quando x = 5 é
## $ 1/\tan(f'(5)) $

```python
b2_a = sy.sin(x**2)
b2_b = sy.cos(2*x)
b2_aD = sy.cos(x**2)*2*x
b2_bD = -2*sy.sin(2*x)
B2Derivative = b2_a*b2_bD + b2_b*b2_aD
B2Derivative.subs(x,5)
```

```python
1/sy.tan(B2Derivative.subs(x,5))
```

```python
Q1_3exp = (x**2)*sy.exp(-x)*sy.log(x)
sy.plot(Q1_3exp, label='(x^2)(e^(-x))(lnx)',legend = True)
```

## (B3):
## $ f(x) = x^2e^{-x}\ln(x) $
## $ g(x) = x^2 $
## $ h(x) = e^{-x}$
## $ z(x) = \ln(x)$
## $ g'(x) = 2x$
## $ h'(x) = -e^{-x}$
## $ z'(x) = 1/x$
## $ f'(x)= -x^2e^{-1}\ln(x) + 2xe^{-x}\ln(x) + (x^2e^{-x})/x $
## $ 1/\tan(f'(5))$

```python
b3_a = x**2
b3_b = sy.exp(-x)
b3_c = sy.log(x)
b3_aD = 2*x
b3_bD = -1*(sy.exp(-x))
b3_cD = 1/x
B3Derivative = b3_aD*b3_b*b3_c + b3_bD*b3_a*b3_c + b3_cD*b3_b*b3_a
display(B3Derivative)
```

```python
B3Derivative.subs(x,5)
```

```python
1/sy.tan(B3Derivative.subs(x,5))
```
